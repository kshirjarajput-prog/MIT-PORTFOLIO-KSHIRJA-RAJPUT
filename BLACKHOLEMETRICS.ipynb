THE SCHWARZSCHILD METRIC

```python

from sympy import symbols, cos, sqrt, pi

r,M = symbols('r,M')

S = sqrt(1 - (2*M) / r)

S
```

```python
M_val = 100
```

```python
import random

def sample():
    return {
        M: M_val,
        r: random.uniform (300,600),
    }
```

```python
data = []
for _ in range(100):
    vals = sample()
    value = S.subs(vals).evalf()
    data.append((vals, value))
```

```python
import pandas as pd

df = pd.DataFrame([list(vals.values()) + [value] for vals, value in data])

df.columns = list(data[0][0].keys()) + ['value']

df.to_csv('Schwarzschild.csv', index=False)
```

THE KERR METRIC

```python
from sympy import symbols, cos, sqrt, pi
import pandas as pd

r,M,a,x = symbols('r,M,a,x')

p = r**2 + a**2*cos(2*x)

K = sqrt(1 - (2*M*r)/p)

K
```

```python
# Fixed mass
M_val = 100
```

```python
import random

def sample():
    return {
        M: M_val,
        a: random.uniform(0,200),
        x: random.uniform(0,float(pi)),
        r: random.uniform(300, 600)
    }
data_K = []
```

```python
data_K = []

for _ in range(100):
    vals = sample()
    value = K.subs(vals).evalf()
    data_K.append((vals, value))
```

```python
import pandas as pd

df = pd.DataFrame([list(vals.values()) + [value] for vals, value in data_K])

df.columns = list(data_K[0][0].keys()) + ['value']

df.to_csv('Kerr.csv', index=False)
```

KERR-NEWMAN METRIC

```python
from sympy import symbols, cos, sqrt, pi
import pandas as pd

r,M,a,x,Q = symbols('r,M,a,x,Q')

N = sqrt((r**2-2*M*r+a**2+Q**2)/(r**2+a**2*(cos(x))**2))


N
```

```python
M_val = 100
```

```python
import random

def sample():
    return {
        M: M_val,
        r: random.uniform(300,600),
        a: random.uniform(1,200),
        Q: random.uniform(1,100),
        x: random.uniform(0,float(pi))
        
    }
data_N = []
```

```python
data_N = []

for _ in range(100):
    vals = sample()
    value = N.subs(vals).evalf()
    data_N.append((vals, value))
```

```python
import pandas as pd

df = pd.DataFrame([list(vals.values()) + [value] for vals, value in data_N])

df.columns = list(data_N[0][0].keys()) + ['value']

df.to_csv('Newman.csv', index=False)
```
